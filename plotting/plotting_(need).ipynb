## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')   # Generates a plot of x vs y_1, using a red dashed line labelled 'Y values'
plt.xlabel('x-values')                                                # Sets the x axis label as 'x-values'
plt.ylabel('y-values')                                                # Sets the y axis label as 'y-values'
plt.title('X vs Y')                                                   # Sets the title of the plot to 'X vs Y'
plt.grid(alpha=.25)                                                   # Generates the grid with the given alpha (opacity)
plt.legend(loc='upper left')                                          # Generates the legend at the given location (upper left)
plt.show()                                                            # Shows the plot
```

### 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y1 values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()                                                      # Default to 'upper left'
plt.show()
```

### 1.7 We need another axis!

```python
x = np.linspace(0, 6, 100)                     
# splits the range of 0 to 6 into 100 equal intervals, plotting a point at each interval, effectively 'smoothing' the curve.
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()                                                   # two different y-plots sharing the same x axis
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')


plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], ls='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.linspace(0,50,500)
y = np.sin(x) * np.e ** (-x/10)

plt.plot(x, y)
plt.xlabel('x', fontsize=15, color='darkblue')
plt.ylabel('y', fontsize=15, color='darkblue')
plt.grid(alpha=0.25)
plt.title(r'$y = e^{-x/10}sin(x)$', fontsize=20)
plt.ylim([-1,1])

plt.show()
```
