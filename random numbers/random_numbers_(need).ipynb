### 1.3 A simple graphical test

```python
from matplotlib import pyplot as plt

n = 1000000
random_numbers = np.random.rand(n)

fig, ax = plt.subplots(nrows=1, ncols=2)

axis = ax[0]
axis.hist(random_numbers, bins=100, alpha=.25)
axis.set_xlabel("Value of random number")
axis.set_ylabel("Frequency")
# For the histogram, as n approaches a large number,
# the histogram bars will all have the same frequency,
# meaning that there is an even distribution for each value.

axis = ax[1]
axis.scatter(range(n), random_numbers, alpha=.25)
axis.set_xlabel("Position in the random number list")
axis.set_ylabel("Value of random number")
# For the scatter plot, as n approaches a large number,
# the scatter points will fill every point in the scatter plot
# meaning that each value has an equal chance to appear at any position
# in the randomised list.

plt.tight_layout()
plt.show()
```

## 2 Simulating a Magic-8 Ball

```python
def shake_magic_8():
    '''
    Function to simulate a Magic-8 ball!
    '''
    options = ['It is certain.', 'It is decidedly so.',
               'Without a doubt.', 'Yes definitely.',
               'You may rely on it.', 'As I see it, yes.',
               'Most likely.', 'Outlook good.',
               'Yes.', 'Signs point to yes.',
               'Reply hazy, try again.', 'Ask again later.',
               'Better not tell you now.', 'Cannot predict now.',
               'Concentrate and ask again.', 'Don\'t count on it.',
               'My reply is no.', 'My sources say no.',
               'Outlook not so good.', 'Very doubtful.']
    return np.random.choice(options)
```

```python
questions = ['Will I pass this module?',
             'Will I survive university?',
             'Will I need therapy?']

for question in questions:
    print(f'Q: {question}')
    print(f'A: {shake_magic_8()}\n')
```

## 3 Flipping Coins

### 3.1 A fair coin

```python
no_of_flips = 10000
result = np.random.choice(['Heads', 'Tails'], no_of_flips)

headno = 0
tailno = 0

for i in result:
    if i == 'Heads':
        headno += 1
    elif i == 'Tails':
        tailno += 1
        
print(f'Number of Heads = {headno}')
print(f'Number of Tails = {tailno}')
```

```python
def flip_coins(no_of_coins=1, probability=.5):
    '''
    Returns the number of values greater that 
    `probability` (considered as 'Heads').
    '''
    results = np.random.rand(no_of_coins)
    no_of_heads = np.sum(results > probability)
    return no_of_heads
```

```python
no_of_flips = 100000000
no_of_heads = flip_coins(no_of_flips)
print(f'Number of heads: {no_of_heads/no_of_flips*100:.2f}%')
```

### 3.2 A biased coin

```python
no_of_coins = 1000000
no_of_heads = flip_coins(no_of_coins, probability = .3)
print(f'Number of heads: {no_of_heads/no_of_coins*100:.2f}%')
```

### 3.3 A flipping experiment

```python
no_of_repeats = 10_000
result = [flip_coins(no_of_coins=10) for _ in range(no_of_repeats)]
no_of_heads, heads_counts = np.unique(result, return_counts=True)

heads_probability = heads_counts/no_of_repeats
```

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.bar(no_of_heads, heads_counts);
plt.xlabel('No. of heads')
plt.ylabel(f'Frequency out of {no_of_repeats}')
plt.table([no_of_heads, heads_counts, heads_probability],
          rowLabels=['No. of Heads', 'Frequency', 'PRNG Probability'],
          loc='bottom',
          bbox=[0, -0.5, 1, 0.3])
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import binom

binom.pmf(k=7, n=10, p=.5)  
# probability of getting 7 heads when 10 fair coins are tossed.
```

```python
binomial_probabilities = binom.pmf(k=no_of_heads, n=10, p=.5)

plt.rcParams['figure.figsize'] = (10,5)
plt.bar(no_of_heads, heads_counts);
plt.xlabel('No. of heads')
plt.ylabel(f'Frequency out of {no_of_repeats}')
plt.table([no_of_heads, heads_counts, heads_probability, np.round(binomial_probabilities,4)],
          rowLabels=['No. of Heads', 'Frequency', 'PRNG Probability','Theo. Probability'],
          loc='bottom',
          bbox=[0, -0.5, 1, 0.3])
plt.tight_layout()
plt.show()
```

## Exercise 1 :  12C, 13C

```python
n = 100000
C13_result = []

for _ in range(n):
    C13_single = np.sum(np.random.rand(27)<0.0111)
    C13_result.append(C13_single) 
# np.sum(np.random.rand(27)<0.0111) refers to the number of C13 atoms in one randomly chosen molecule. 
# The for loop repeats the randomisation to produce a list consisting of the individual numbers of C13 atoms 
# in n number of molecules.

no_of_C13, C13_count = np.unique(C13_result, return_counts = True)
# The unique number of C-13 present is recorded, and the number of times each unique number appears is counted.

C13_probability = C13_count/n
# i.e., number of molecules with i number of C-13, out of all molecules checked. (i >= 0)

mean = np.mean(C13_result)
std = np.std(C13_result)
# calculates the mean and standard deviation of the number of C-13.

plt.rcParams['figure.figsize'] = (10,5)
plt.bar(no_of_C13, C13_count);
plt.xlabel('No. of C-13')
plt.ylabel(f'Number of molecules out of {n}')
plt.text(3,50000,f'Mean = {np.round(mean,4)} \nStandard Deviation = {np.round(std,4)}', fontsize = 10)
plt.table([no_of_C13, C13_count, C13_probability],
          rowLabels=['No. of C13', 'Number of molecules', 'Probability'],
          loc='bottom',
          bbox=[0, -0.5, 1, 0.3])
plt.tight_layout()
plt.show()
```
