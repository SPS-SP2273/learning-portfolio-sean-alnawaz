```python
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:
        os.mkdir(path)
        print(f'Creating {path}')
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')
```

```python
for person in ['John', 'Paul', 'Ringo','Jake']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

### 2.5 Copying files

```python
for person in ['John','Paul','Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('sp2273_logo.png', path_to_destination)
    print(f'Copied file to {path_to_destination}')
```

```python
for person in ['John','Paul','Ringo']:
    path_to_imgs = os.path.join('people', person,'imgs')
    if not os.path.exists(path_to_imgs):
        os.mkdir(path_to_imgs)
        
    current_path_of_logo = os.path.join('people', person, 'sp2273_logo.png')
    new_path_of_logo = os.path.join('people', person, 'imgs','sp2273_logo.png')
    
    shutil.move(current_path_of_logo, new_path_of_logo)
    print(f'Moved logo to {new_path_of_logo}')
```

## Exercise 8 :  Tidying a collaboration

```python
os.chdir('..\os-collaboration-exercise-data')
```

```python
glob.glob('*singapore.txt')      
```

```python
months = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

path = os.path.join('oct-08_singapore.txt')
checkexist = os.path.exists(path)                 # Checks if file with given name and extension exists
if checkexist == True:
    newpath = path.replace('_','-')               # Replaces _ into -
    splitpath = newpath.split('-')                # Splits the path name into a list, using - as the separator
    splitpath[0] = months.get(splitpath[0])       # Replaces key with value
    finalpath = '-'.join(splitpath)               # Recombines the list into the final path name
    os.rename(path,finalpath)                     # Renames the file
    print(f'{path} ----> {finalpath}')            
    
else:
    print(f'{path} does not exist or has already been renamed!')     
```

```python
def rename_my_file(old_filename):
    path = os.path.join(old_filename)
    checkexist = os.path.exists(path)
    if checkexist == True:
        newpath = path.replace('_','-')
        splitpath = newpath.split('-')
        splitpath[0]=months.get(splitpath[0])
        finalpath = '-'.join(splitpath)
        os.rename(path,finalpath)
        print(f'{old_filename} ----> {finalpath}')
        
    else:
        print(f'{old_filename} does not exist or has already been renamed!')
```

```python
filelist = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for i in filelist:
    rename_my_file(i)
```

```python
for j in glob.glob('*_*.txt'):
    rename_my_file(j)
```

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila', 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

[shutil.move(p, q) for q in cities for p in glob.glob(f'*{q.lower()}.*')]

# A nested for loop, which moves files with the city in their name to the respective matching folders.
```
